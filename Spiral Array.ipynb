# [ Conclusion ]
* Pattern found: 
  - 조각(segment)의 길이는 일반적으로 l, s 두가지로 정의될 수 있다.
     - l == s (square)
     - l != s (rectangle)
  - 조각은 ...-l-s-l-s-l-... 순으로 번갈아가며 나아간다.
  - 반복횟수는 3 -> 2 -> 2 -> ... -> 2 -> (1) 순서로 줄어든다. 
  - 다음 반복횟수로 넘어갈 때마다 l-=1; s-=1 을 취한다.
  - l 혹은 s 둘중 하나라도 0이 될때를 end point로 정의하며, 그 때는 그 0을 1로 치환후 끝낸다.
  - 시작 조각이 l인지 s인지는 회전방향과 초기위치로 결정된다.

# 2. Define `class`: `seg_path`

매번 클래스 연습해야지 해야지 하다가 드디어 고생 끝에 시도해봤습니다.

```python
class seg_path:
    def __init__(self, init_idx, sprial):
        self.r_pos = init_idx[0]
        self.c_pos = init_idx[1]
        self.sprial = sprial
        self.get_dir()
    
    def sym_turn(self):
        seg_lst = ["r","c"]
        seg_lst.remove(self.sym)
        self.sym = seg_lst[0]
    
    def dir_turn(self):
        self.dir_cmpl = self.dir_cmpl * 1j
        self.dir = (self.dir_cmpl).real

    def get_dir(self):
        # get back init_idx
        idcs = [self.r_pos, self.c_pos]

        # check which edge 'init_idx' locates
        # [True, True]  : "ul"
        # [True, False] : "ur"
        # [False, True] : "bl"
        # [False, False]: "br"
        check_idx = list(map(lambda x, y: x==y, idcs, [0,0]))

        # Assume sprial=="clock"
        # Complex directions (pattern found empirically)
        """
 - clock-wise ("clock")
 |        (+1)
 |         -->
 |      ^        |  (+1)
 | (-1) |        v
 |         <--
 |         (-1)
 |
 |- init_pos (init_idx)
 |    |- "ul" ([0,0]): init_dir_cmpl = +1 (-> +1)  >>  1-1j
 |    |- "ur" ([0,c]): init_dir_cmpl = +1 (-> -1)  >>  1+1j
 |    |- "bl" ([r,0]): init_dir_cmpl = -1 (-> +1)  >> -1-1j
 |    |- "br" ([r,c]): init_dir_cmpl = -1 (-> -1)  >> -1+1j
 |
 |- init_dir = init_dir_cmpl.real()
 |
 - Anti clock-wise ("anti")
 |        (-1)
 |         <--
 | (+1) |        ^
 |      v        |  (-1)
 |         -->
 |         (+1)
 |- init_pos
 |    |- "ul" ([0,0]): init_dir_cmpl = +1 (-> +1)  >>  1-1j
 |    |- "ur" ([0,c]): init_dir_cmpl = -1 (-> +1)  >> -1-1j
 |    |- "bl" ([r,0]): init_dir_cmpl = +1 (-> -1)  >>  1+1j
 |    |- "br" ([r,c]): init_dir_cmpl = -1 (-> -1)  >> -1+1j
 |
 |- init_dir = init_dir_cmpl.real()
 |
 - "clock" <-> "anti" : conversion by "init_dir.conjugate()*-1*1j"
        """
        if   check_idx == [True, True]:
            self.dir_cmpl =  1-1j
            self.sym = "c"
        elif check_idx == [True, False]:
            self.dir_cmpl =  1+1j
            self.sym = "r"
        elif check_idx == [False, True]:
            self.dir_cmpl = -1-1j
            self.sym = "r"
        elif check_idx == [False, False]:
            self.dir_cmpl = -1+1j
            self.sym = "c"
        
        # check 'sprial'
        if self.sprial == "clock":
            pass
        elif self.sprial == "anti":
            self.dir_cmpl = self.dir_cmpl.conjugate()*(-1)*(1j) # emprical law
            self.sym_turn()
        else:
            raise ValueError("Wrong 'sprial'. Choose one of 'clock' or 'anti'.")
        
        # Also store real direction
        self.dir = (self.dir_cmpl).real
        
    def get_path(self, syms):
        r = syms["r"]
        c = syms["c"]

        if   self.sym == "c":
            if   self.dir ==  1:
                path = (self.r_pos, range(self.c_pos,self.c_pos+c))
                self.c_pos += c # Update c_pos
            elif self.dir == -1:
                path = (self.r_pos, range(self.c_pos, self.c_pos-c, -1))
                self.c_pos -= c # Update c_pos

        elif self.sym == "r":
            if   self.dir ==  1:
                path = (range(self.r_pos,self.r_pos+r), self.c_pos)
                self.r_pos += r # Update r_pos
            elif self.dir == -1:
                path = (range(self.r_pos, self.r_pos-r, -1), self.c_pos)
                self.r_pos -= r # Update r_pos
        
        self.path = path
        
        # For next turn
        self.sym_turn()
        self.dir_turn()
```

```python
def get_init_pos_tup(input_dim):
    """
 - Initial position:
    |- r = input_dim[0] - 1
    |- c = input_dim[1] - 1 
    |============================|
    | edge               | index |
    |--------------------|-------|
    | "ul": upper-left   | [0,0] |
    | "ur": upper-right  | [0,c] |
    | "bl": bottom-left  | [r,0] |
    | "br": bottom-right | [r,c] |
    |============================|
    """
    init_pos_tup = dict(ul = [0,0], 
                        ur = [0,input_dim[1]-1], 
                        bl = [input_dim[0]-1,0], 
                        br = [input_dim[0]-1, input_dim[1]-1])
    return(init_pos_tup)
```

# 3. Define `main` function

```python
def main(input_dim, input_pos, input_dir, verbose=True):
    import numpy as np
    """
     - Grid dimension: 2D with any number
        e.g.) [5,5] , [6,6] , [3,4]

     - Initial position: "ul", "ur", "bl", "br"
        c.f.)
        |- r = input_dim[0] - 1
        |- c = input_dim[1] - 1 
        |============================|
        | edge               | index |
        |--------------------|-------|
        | "ul": upper-left   | [0,0] |
        | "ur": upper-right  | [0,c] |
        | "bl": bottom-left  | [r,0] |
        | "br": bottom-right | [r,c] |
        |============================|

    - Sprial directions: "clock", "anti"
         2 Possible direction: clock-wise (clock), anticlock-wise (anti)
    
    """

    # Initialize grid with zeros
    grid = np.zeros(input_dim)
    
    # Length of rows and columns
    row_len = input_dim[0]
    col_len = input_dim[1]
    
    # Generate straight sequence with respect to the given dimension.
    # e.g.)
    #   6 x 6 = 36 -> [ 0, 1, 2, ..., 34, 35]
    #   3 x 4 = 12 -> [ 0, 1, 2, ..., 10, 11]
    straight_seq = np.array(range(row_len * col_len))
    init_pos_tup = get_init_pos_tup(input_dim)
    init_idx = init_pos_tup[input_pos]
    
    # Initialize 
    r, c = row_len-1, col_len-1 # row index, column index
    straight_seq_copy = straight_seq.copy()
    
    # Start
    seg = seg_path(init_idx, input_dir)

    # Stage iteration
    stage_num = 0
    while True:
        # Stage number
        stage_num += 1

        # Printing
        if verbose:
            print("|> Stage "+str(stage_num))

        # Step number
        if stage_num == 1:
            step_num = 3
        else:
            step_num = 2

        # Step iteration
        for step in range(step_num):
            # End point
            if   r == 0 and c == 0:
                r = 1
                c = 1
            elif r == 0:
                r = 1
            elif c == 0:
                c = 1
            else:
                pass

            # Printing
            if verbose:
                print("|-> Step "+str(step+1))

            # Get segment index
            idx_dir = range(0,eval(seg.sym))

            # Get segment values from sequence
            seg_vals = straight_seq_copy[idx_dir]

            # Get path to insert segment values
            syms = {"r":r, "c":c}
            seg.get_path(syms)

            # Insert segment values into the path
            grid[seg.path] = seg_vals

            # Delete used segment from original sequence
            straight_seq_copy = np.delete(straight_seq_copy, idx_dir)

            # Printing
            if verbose:
                print(grid)

            # Break for loop
            if len(straight_seq_copy) == 0:
                break
        
        # Break while loop
        if len(straight_seq_copy) == 0:
            break
        
        # Decrease r & c
        r-=1; c-=1
    
    if verbose:
        print("|> Filling Spiral Array Done!")

    return(grid)
```
